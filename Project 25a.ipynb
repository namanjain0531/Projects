# HANDLING THE OUTLIERS

```python
from scipy.stats import zscore
z=np.abs(zscore(bike))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
bike_new=bike[(z<3).all(axis=1)]
bike_new
```

# DIVIDE THE DATA INTO X AND Y

```python
x=bike_new.drop("cnt",axis=1)
x
```

```python
y=bike_new['cnt']
y
```

# SPLITTING THE DATA INTO TESTING AND TRAINING PHASE

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# APPLY MULTIPLE ALGORITHMS

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
predlr=lr.predict(x_test)
print(r2_score(y_test,predlr))
print(mean_squared_error(y_test,predlr))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kfold=KFold(n_splits=10, random_state=0)
model=LinearRegression()
print(cross_val_score(model, x, y, cv=kfold))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(max_depth=4,
                            min_samples_split=5,
                            max_leaf_nodes=10)
dtr.fit(x,y)
```

```python
dtr.score(x,y)
```

```python
#save the model
import warnings
warnings.filterwarnings('ignore')
from sklearn.externals import joblib
joblib.dump(lr,"bike_sharing.pkl")
```
