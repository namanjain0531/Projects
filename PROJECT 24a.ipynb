# DIVIDE THE DATA INTO X AND Y

```python
x=house.drop(10,axis=1)
x
```

```python
y=house[10]
y
```

# SPLIT THE DATA IN TRAINING AND TESTING PHASE

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# SCALING THE DATA

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(x)
```

# APPLY MODEL AND ALGORITHMS

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=300, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(estimator=rfc, X=x_train, y=y_train, cv=5))
```

```python
from sklearn.model_selection import GridSearchCV
```

# HYPERPARAMETERS TO TUNE

```python
grid_param={
    'n_estimators': [100, 300, 500, 800, 1000],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}
```

```python
gsv= GridSearchCV(estimator=rfc,
                     param_grid=grid_param,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
```

```python
gsv.fit(x_train,y_train)
```

```python
best_parameters=gsv.best_params_
print(best_parameters)
```

```python
best_score=gsv.best_score_
print(best_score)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
predlg=lg.predict(x_test)
print(accuracy_score(y_test,predlg))
print(confusion_matrix(y_test,predlg))
print(classification_report(y_test,predlg))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,predknn))
print(classification_report(y_test,predknn))
```
